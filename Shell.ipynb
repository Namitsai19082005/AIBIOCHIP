```python
%%shell
range=20
for i in $(seq 1 $range); do
    count=0
    for j in $(seq 1 $i); do
       if((i%j==0)); then
          count=$((count+1));
          fi
    done
    if((count==2)); then
        echo $i
        fi
done
```
